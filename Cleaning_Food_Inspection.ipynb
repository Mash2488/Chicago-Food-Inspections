### Filling Missing Facility Types Based on Known Business Names

* To impute missing `facility_type` values, I first reviewed the most common `aka_name` values among those rows. This revealed well-known chains such as SUBWAY, 7-ELEVEN, and DUNKIN DONUTS, where the expected facility type was clear.

* Based on these insights, I created a mapping dictionary to assign the most likely `facility_type` to these businesses. Remaining missing values were filled with `'Unknown'`.


```python
# Filling missing facility types by mapping known AKA names to their most likely category

known_facility_types = {
    '7-ELEVEN': 'Grocery Store',
    'SUBWAY': 'Restaurant',
    'DUNKIN DONUTS': 'Restaurant',
    'WHITE HEN PANTRY': 'Grocery Store',
    'MIDWEST BUDDHIST TEMPLE': 'Religious Organization'
    
}
```

```python
# Filling missing facility types by mapping known aka_names using the dictionary created earlier.
mask = df['facility_type'].isna()
df.loc[mask, 'facility_type'] = df.loc[mask, 'aka_name'].map(known_facility_types)
```

```python
# Filling remaining missing facility types with unknown
df['facility_type'] = df['facility_type'].fillna('Unknown')
```

### Interpreting the risk column:
* Risk 1 (High): Higher likelihood of affecting public health

* Examples: restaurants, delis, schools

* Risk 2 (Medium): Moderate risk

* Examples: grocery stores, bakeries

* Risk 3 (Low): Lower risk

* Examples: warehouses, gas stations, vending machines

```python
# Investigating mssing risk values
df['risk'].value_counts(dropna=False)
```

```python
# Viewing overall risk distribution by facility type to identify dominant risk levels.
df.groupby('facility_type')['risk'].value_counts(normalize=True).unstack().fillna(0).style.format("{:.0%}")
```

```python
# Checking risk distribution for facility types with missing risk values to guide imputation strategy.
top_missing_types = df[df['risk'].isna()]['facility_type'].value_counts().index.tolist()

df[df['facility_type'].isin(top_missing_types)].groupby('facility_type')['risk'].value_counts(normalize=True).unstack().fillna(0).style.format("{:.0%}")
```

```python
# These facility types were chosen because they appeared most frequently among rows with missing risk values
# And had a strong majority (typically over 70%) mapped to a specific risk level.
# This allows for confident imputation without introducing bias.

risk_by_facility = {
    'Restaurant':'Risk 1 (Hight)',
    'AFTER SCHOOL PROGRAM':'Risk 2 (Medium)',
    'CHURCH':'Risk 2 (Medium)',
    'Illegal Vendor':'Risk 3 (Low)',
    'Mobile Food Dispensor':'Risk 3 (Low)',
    'RETAIL':'Risk 3 (Low)',
    'Unknown':'Risk 3 (Low)', # for facility_type == 'Unknown'
    'WAREHOUSE':'Risk 1 (High)',
    'Wholesale':'Unknown'
}
```

```python
# Filling missing risk values based on most common risk level for each facility type.
missing_risk_mask = df['risk'].isna()
df.loc[missing_risk_mask, 'risk'] = df.loc[missing_risk_mask, 'facility_type'].map(risk_by_facility)
```

```python
#Assigning remaining missing risks with unknown
df['risk'] = df['risk'].fillna('Unknown Risk')
```

```python
# Capitalizing CHICAGO for consistency and removing known typo's
df['city'] = df['city'].str.upper().str.strip()
df['city'] = df['city'].replace({'CCHICAGO': 'CHICAGO'})
```

```python
#If city and state is IL then it is almost certainly chicago
df.loc[df['city'].isna() & (df['state'] == 'IL'), 'city'] = 'CHICAGO'

# Fill state if city is CHICAGO
df.loc[df['state'].isna() & (df['city'] == 'CHICAGO'), 'state'] = 'IL'
```
